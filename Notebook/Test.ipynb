```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.utils import shuffle
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, accuracy_score, classification_report, confusion_matrix
```

```python
loaded_bundle = joblib.load(r'D:\Git\DSS\Model\Decision_tree_Esemble.pkl')
```

```python
loaded_bundle
```

```python
test=pd.read_csv('Data/Train.csv')
```

```python
feature={'is_purchased','avg_price','max_price','user_session'}
X_test=test.drop(columns=feature)
y_test=test['is_purchased']
```

```python
loaded_models = loaded_bundle['model']
loaded_threshold = loaded_bundle['threshold']
```

```python
all_predictions_new = []
for model in loaded_models:
    all_predictions_new.append(model.predict_proba(X_test)[:, 1])

y_prob_new_ensemble = np.mean(all_predictions_new, axis=0)
y_pred_new_final = (y_prob_new_ensemble >= loaded_threshold).astype(int)
```

```python
print(classification_report(y_test, y_pred_new_final))
```
